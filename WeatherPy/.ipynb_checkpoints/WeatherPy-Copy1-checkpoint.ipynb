```python
# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json

from datetime import datetime


# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
 # after ? is the info that im retrieving 
url= "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="

#query_url

# Temperature (F) vs. Latitude
# Humidity (%) vs. Latitude
# Cloudiness (%) vs. Latitude
# Wind Speed (mph) vs. Latitude

setcount=1
citycount=0 
date = []
country = []
name = []
temp = []
lat = []
humid = []
cloud = []
wind = []
lon = []



# # # Loop through the list of cities and perform a request for data on each
for city in cities:
    
    if citycount == 50:
        citycount = 0 
        setcount +=1
    
    try:
        #print(query_url+city)
        response = requests.get(query_url + city).json()
        
        date.append(response['dt'])
        country.append(response['sys']['country'])
        lat.append(response['coord']['lat'])
        temp.append(response['main']['temp'])
        lon.append(response['coord']['lon'])
        humid.append(response['main']['humidity'])
        cloud.append(response['clouds']['all'])
        wind.append(response['wind']['speed'])
        name.append(city)
        citycount +=1
        print(f"Processing record {citycount} of {setcount}:{city}")
    except:
        print(f"No data for {city}")



# query_url
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_df= pd.DataFrame({"City":name,
                        "Country": country,
                        "Max Temp": temp,
                        "Humidity": humid,
                        "Cloudiness": cloud,
                        "Wind Speed": wind,
                        "Lat": lat,
                        "Lng":lon,
                        "Date": date
                        })

weather_df
```

```python
# Save the dataframe into a csv file

weather_df.to_csv("weather.csv",encoding="utf-8", index=False)
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
# set axis
xlabel= City_Data['Lat'] 
ylabel= City_Data['Cloudiness']


(slope, intercept, rvalue, pvalue, stderr) = linregress(xlabel, ylabel)
regress_values = xlabel * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))


plt.scatter(xlabel, ylabel)

plt.plot(xlabel,regress_values,"r-")

plt.annotate(line_eq,(77,38),fontsize=15,color="red")

plt.title(f"Latitude vs Cloudiness Linear Regression {dateon}")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness")

print(f"R squared: {rvalue**2}")
plt.savefig(f'../output_data/Latitude vs Cloudiness Linear Regression on {dateon}.png')
plt.show()
```
